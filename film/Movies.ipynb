```python
# Banyak data kosong

NaN = df.isna().sum()

print(NaN)
```

```python
# Diagram Lead Studio
df["Lead Studio"].value_counts().plot(kind = "barh")
```

```python
# Histogram score Rotten Tomatoes
df[["Rotten Tomatoes"]].plot(kind="hist", bins=10)

# Skor Terendah Rotten Tomatoes
tomato_min = df["Rotten Tomatoes"].min()
print(f"Skor terendah Rotten Tomatoes : {tomato_min}")

# Skor Tertinggi Rotten Tomatoes
tomato_max = df["Rotten Tomatoes"].max()
print(f"Skor terendah Rotten Tomatoes : {tomato_max}")

dfmin = df.loc[df["Rotten Tomatoes"] == tomato_min] # data frame dengan data Rotten Tomatoes min
dfmax = df.loc[df["Rotten Tomatoes"] == tomato_max] # data frame dengan data Rotten Tomatoes max

pd.concat([dfmin, dfmax]) # menggabungkan 2 data frame
```

```python
# Histogram score Audiens
df[["Audience Score"]].plot(kind="hist", bins=10)

# Skor Terendah Audience Score
audience_min = df["Audience Score"].min()
print(f"Skor terendah Audience Score : {audience_min}")

# Skor Tertinggi Audience Score
audience_max = df["Audience Score"].max()
print(f"Skor terendah Audience Score : {audience_max}")

dfmin = df.loc[df["Audience Score"] == audience_min] # data frame dengan data Audience Score min
dfmax = df.loc[df["Audience Score"] == audience_max] # data frame dengan data Audience Score max

pd.concat([dfmin, dfmax]) # menggabungkan 2 data frame
```

```python
# Data   : Film Hollywood (score, genre, profit)
# Source : https://github.com/gchan/hollywood-budgets/blob/master/public/data/Most%20Profitable%20Hollywood%20Stories.csv
# Format : .csv
# Ukuran : 641 baris x 15 kolom, 62.127 bytes

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("CleanedMovie.csv", sep=";")
```

```python
# Preview Data 

display(df[:5])

# Banyak Data 

print(f"Banyaknya Data : {len(df)}")
```

```python
# Atribut Data
                        
#                      LABEL                                                                DESKRIPSI                                        JENIS DATA   
# 1. Film                                               : Judul film                                                                       : -           
# 2. Lead Studio                                        : Studio film                                                                      : Kategorikal 
# 3. Rotten Tomatoes                                    : Score film menurut https://www.rottentomatoes.com/                               : Kuantitatif 
# 4. Audience Score                                     : Score film menurut penonton                                                      : Kuantitatif 
# 5. Story                                              : Tipe cerita                                                                      : Kategorikal 
# 6. Genre                                              : Genre film                                                                       : Kategorikal 
# 7. Number of Theatres in US Opening Weekend           : Banyaknya bioskop yang menayangkan film pada opening weekend                     : Kuantitatif 
# 8. Box Office Average per US Cinema (Opening Weekend) : Rata-rata keuntungan kotor bioskop lokal pada opening weekend (dalam dolar)      : Kuantitatif 
# 9. Domestic Gross                                     : Total keuntungan kotor bioskop lokal (US) (dalam juta dolar)                     : Kuantitatif 
# 10. Foreign Gross                                     : Total keuntungan kotor bioskop luar negeri (dalam juta dolar)                    : Kuantitatif 
# 11. Worldwide Gross                                   : Total keuntungan kotor (dalam juta dolar)                                        : Kuantitatif
# 12. Budget                                            : Budget film (dalam juta dolar)                                                   : Kuantitatif 
# 13. Profitability                                     : Rasio keuntungan dengan modal                                                    : Kuantitatif 
# 14. Opening Weekend                                   : Total keuntungan kotor pada opening weekend (dalam juta dolar)                   : Kuantitatif  
# 15. Year                                              : Tahun rilis                                                                      : Kategorikal 
```

```python
# Karakteristik Data

df.describe()
```

```python
# Karakteristik Data

list_p10 = []
list_p90 = []

list_p10.append("10%")
list_p90.append("90%")

atribut = ["Rotten Tomatoes", "Audience Score", "Number of Theatres in US Opening Weekend",	"Box Office Average per US Cinema (Opening Weekend)", "Domestic Gross",	"Worldwide Gross", "Budget", "Opening Weekend",	"Year"]

for i in atribut:
    p10 = df[f"{i}"].quantile(0.1)
    p90 = df[f"{i}"].quantile(0.9)

    list_p10.append(p10)
    list_p90.append(p90)

atribut = ["","Rotten Tomatoes", "Audience Score", "Number of Theatres in US Opening Weekend",	"Box Office Average per US Cinema (Opening Weekend)", "Domestic Gross",	"Worldwide Gross", "Budget", "Opening Weekend",	"Year"]


p10_90= pd.DataFrame([atribut, list_p10, list_p90])

display(p10_90)


```

```python
# Banyak Data Kosong

list_kolom = []
list_data_kosong = []
list_persen_kosong = []

for kolom in df.columns:

    data_kosong = df[kolom].isna().sum()

    persen_kosong = f"{round(data_kosong/len(df)*100,2)}%"

    list_kolom.append(kolom)

    list_data_kosong.append(f"{data_kosong}/{len(df)}")

    list_persen_kosong.append(persen_kosong)

    # print(kolom, data_kosong, persen_kosong)

NaN = pd.DataFrame({"Atribut":list_kolom, "Banyak Data Kosong":list_data_kosong, "Persentase Data Kosong":list_persen_kosong})

display(NaN)
```

```python
# 3. Rotten Tomatoes
df[["Rotten Tomatoes"]].plot(kind="hist", bins=10)

# Skor Terendah Rotten Tomatoes
tomato_min = df["Rotten Tomatoes"].min()
print(f"Skor terendah  : {tomato_min}")

# Skor Tertinggi Rotten Tomatoes
tomato_max = df["Rotten Tomatoes"].max()
print(f"Skor tertinggi : {tomato_max}")

dfmin = df.loc[df["Rotten Tomatoes"] == tomato_min] # data frame dengan data Rotten Tomatoes min
dfmax = df.loc[df["Rotten Tomatoes"] == tomato_max] # data frame dengan data Rotten Tomatoes max

pd.concat([dfmin, dfmax]) # menggabungkan 2 data frame
```
